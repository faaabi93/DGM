```python
import pandas as pd
import matplotlib.pyplot as plt

colnames = ["X", "Y", "Z"]
data = pd.read_csv("/Users/fabian.baiersdoerfer/Documents/GitHub/DGM/dgm1000.utm32s.xyzascii/dgm1000/dgm1000_utm32s.xyz", names = colnames, sep=" ")

print(data)


to_drop = []
for index, row in data.iterrows():
    if row["Y"] % 10000 != 0 or row["X"] % 2500 != 0:
        to_drop.append(index)

data.drop(to_drop, inplace=True)
```

```python
pd.options.mode.chained_assignment = None
import numpy as np
import scipy.signal


fig1 = plt.figure(figsize=(30,30))
ax1 = fig1.add_subplot(111)
ax1.set_facecolor('black')


for y in y_vals:
    temp_df = data.loc[data["Y"] == y]
    temp_df["new_Y"] = temp_df["Y"] + 60 * temp_df["Z"]
    # using savgol: window_size must be odd and smaller than x
    # polyorder: must be smaller than window_size:
    if len(temp_df["X"]) > 3:
        polyorder = 2
        window_length = 3
        yhat = scipy.signal.savgol_filter(temp_df["new_Y"], window_length, polyorder) # window size 51, polynomial order 3
        ax1.plot(temp_df["X"], yhat)
    else:
        ax1.plot(temp_df["X"], temp_df["new_Y"])



colormap = plt.cm.autumn #nipy_spectral, Set1,Paired   
colors = [colormap(i) for i in np.linspace(0, 1,len(ax1.lines))]
for i,j in enumerate(ax1.lines):
    j.set_color(colors[i])
```

```python
pd.options.mode.chained_assignment = None
import numpy as np
import numpy.ma as ma
import scipy.signal
import more_itertools as mit


fig1 = plt.figure(figsize=(30,30))
ax1 = fig1.add_subplot(111)
ax1.set_facecolor('black')


for y in y_vals:
    temp_df = data.loc[data["Y"] == y]
    min_x = min(data["X"])
    min_y = min(data["Y"])
    temp_df["new_Y"] = temp_df["Y"] + 60 * temp_df["Z"]
    temp_df["adj_x"] = (temp_df["X"] - min_x) / 5000
    temp_df["adj_y"] = (temp_df["Y"] - min_y) / 5000
    temp_df = temp_df.astype("int32")
    adj_x = temp_df["adj_x"]
    adj_y = temp_df["adj_y"].values.tolist()
    list_to_print = []
    for group in mit.consecutive_groups(adj_x):
        list_to_print.append(list(group))
    plt.plot(adj_x, adj_y)
    count = 0
    count2 = 0
    for lst in list_to_print:
        print(len(lst)
    print("---")


colormap = plt.cm.autumn #nipy_spectral, Set1,Paired   
colors = [colormap(i) for i in np.linspace(0, 1,len(ax1.lines))]
for i,j in enumerate(ax1.lines):
    j.set_color(colors[i])
```
