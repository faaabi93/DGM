```python
to_drop = []
for index, row in data.iterrows():
    if row["Y"] % 25000 != 0 or row["X"] % 2500 != 0:
        to_drop.append(index)
```

```python
data.drop(to_drop, inplace=True)
```

```python
y_vals = data["Y"].unique().tolist()

fig1 = plt.figure(figsize=(30,30))
ax1 = fig1.add_subplot(111)
ax1.set_facecolor('black')


for y in y_vals:
    temp_df = data.loc[data["Y"] == y]
    temp_df["new_Y"] = temp_df["Y"] + 30 * temp_df["Z"]
    ax1.scatter(temp_df["X"], temp_df["new_Y"], color="grey")
    
"""colormap = plt.cm.summer #nipy_spectral, Set1,Paired   
colors = [colormap(i) for i in np.linspace(0, 1,len(ax1.lines))]
for i,j in enumerate(ax1.scatter):
    j.set_color(colors[i])"""
```
